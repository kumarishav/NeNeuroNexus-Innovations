```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop('Class', axis=1),df['Class'], test_size=0.3, random_state=42)
```

```python
print("{} examples of training, {} examples fro testing".format(len(x_train),len(x_test)))
```

```python
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(y_true,y_pred):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', annot_kws={"size":16})
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(max_iter=1000)
logreg.fit(x_train, y_train)
prediction_logreg = logreg.predict(x_test)

plot_confusion_matrix(y_test, prediction_logreg)

plt.title(f'Confusion Matrix for Logistic Regression')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
prediction_rf = rf.predict(x_test)
plot_confusion_matrix(y_test,prediction_rf)

plt.title(f'Confusion Matrix for Random Forest')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
prediction_dt = dt.predict(x_test)
plot_confusion_matrix(y_test,prediction_dt)
plt.title(f'Confusion Matrix for Decision Tree')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(x_train,y_train)
prediction_svm = svm.predict(x_test)
plot_confusion_matrix(y_test,prediction_svm)
plt.title(f'Confusion Matrix for SVM')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

```python
from sklearn.neural_network import MLPClassifier
nn = MLPClassifier(hidden_layer_sizes=(64, 64), max_iter=1000)
nn.fit(x_train,y_train)
prediction_nn = nn.predict(x_test)
plot_confusion_matrix(y_test, prediction_nn)
plt.title(f'Confusion Matrix for Neural Network')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score
def safe_precision_score(y_true, y_pred):
    if sum(y_pred)==0:
        return 0.0
    return precision_score(y_true, y_pred)

models = ['Logistic Regression', 'Random Forest', 'Decision Tree', 'SVM', 'Neural Network']
precision = [safe_precision_score(y_test,prediction_logreg),
            safe_precision_score(y_test,prediction_rf),
            safe_precision_score(y_test,prediction_dt),
            safe_precision_score(y_test,prediction_svm),
            safe_precision_score(y_test,prediction_nn)]
recall = [recall_score(y_test,prediction_logreg),
          recall_score(y_test,prediction_rf),
          recall_score(y_test,prediction_dt),
          recall_score(y_test,prediction_svm),
          recall_score(y_test,prediction_nn)]
f1 = [f1_score(y_test,prediction_logreg),
      f1_score(y_test,prediction_rf),
      f1_score(y_test,prediction_dt),
      f1_score(y_test,prediction_svm),
      f1_score(y_test,prediction_nn)]
roc_auc = [roc_auc_score(y_test, prediction_logreg),
           roc_auc_score(y_test, prediction_rf),
           roc_auc_score(y_test, prediction_dt),
           roc_auc_score(y_test, prediction_svm),
           roc_auc_score(y_test, prediction_nn)]

# Create a DataFrame for easier visualization
result_df = pd.DataFrame({
    'Model': models,
    'Precision':precision,
    'Recall':recall,
    'F1-Score':f1,
    'AUC-ROC': roc_auc
})

# Plot the results using a bar chart
plt.figure(figsize=(12, 8))
sns.barplot(data=result_df, x='Model', y='Precision', color='skyblue', label='Precision')
sns.barplot(data=result_df, x='Model', y='Recall', color='salmon', label='Recall')
sns.barplot(data=result_df, x='Model', y='F1-Score', color='green', label='F1-Score')
sns.barplot(data=result_df, x='Model', y='AUC-ROC', color='purple', label='AUC-ROC')
plt.title('Comparison of Metrics for Different Models')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import roc_curve

# Function to plot the ROC Curve
def plot_roc_curve(y_true, y_scores, label):
    fpr, tpr, thresholds = roc_curve(y_true, y_scores)
    plt.plot(fpr, tpr, label=f'{label} (AUC = {roc_auc_score(y_true, y_scores):.2f})')

# Add ROC Curves for each model
plot_roc_curve(y_test, logreg.predict_proba(x_test)[:, 1], 'Logistic Regression')
plot_roc_curve(y_test, rf.predict_proba(x_test)[:, 1], 'Random Forest')
plot_roc_curve(y_test, dt.predict_proba(x_test)[:, 1], 'Decision Tree')
plot_roc_curve(y_test, svm.decision_function(x_test), 'SVM')
plot_roc_curve(y_test, nn.predict_proba(x_test)[:, 1], 'Neural Networks')

# Layout adjustments
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve for Fraud Detection Models')
plt.legend()
plt.show()
```

```python
# DataFrame to store metrics
metrics_df = pd.DataFrame({
    'Model': models,
    'AUC-ROC': roc_auc,
    'Precision': precision,
    'Recall': recall,
    'F1-Score': f1
})

# Display the metrics DataFrame
print(metrics_df)

# Determine the best model based on a specific metric
best_model_auc = metrics_df.loc[metrics_df['AUC-ROC'].idxmax()]
best_model_precision = metrics_df.loc[metrics_df['Precision'].idxmax()]
best_model_recall = metrics_df.loc[metrics_df['Recall'].idxmax()]
best_model_f1 = metrics_df.loc[metrics_df['F1-Score'].idxmax()]

print("\nBest Model (AUC-ROC):")
print(best_model_auc)

print("\nBest Model (Precision):")
print(best_model_precision)

print("\nBest Model (Recall):")
print(best_model_recall)

print("\nBest Model (F1-Score):")
print(best_model_f1)
```

```python
# Create a bar chart for precision, recall, and F1-Score metrics
plt.figure(figsize=(12, 8))
metrics_df_melted = pd.melt(metrics_df, id_vars='Model', var_name='Metric', value_name='Value')
sns.barplot(data=metrics_df_melted, x='Model', y='Value', hue='Metric', palette='viridis')
plt.title('Comparison of Metrics for Fraud Detection Models')
plt.ylabel('Value')
plt.show()
```
